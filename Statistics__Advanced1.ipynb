Ques3. Explain the difference between discrete and continuous distributions.
Ans:Discrete distributions model countable, distinct outcomes (e.g., coin flips, integers) using probability mass functions, where exact values have specific probabilities. Continuous distributions model uncountable, measurable ranges (e.g., time, weight) using probability density functions, where exact values have zero probability, focusing instead on intervals.
Key Differences at a Glance
.Data Type: Discrete data is counted; Continuous data is measured.
.Values: Discrete values are separate/distinct; Continuous values can be any value within a range.
.Probability Measure: Discrete uses Probability Mass Function (PMF); Continuous uses Probability Density Function (PDF).
.Graphing: Discrete distributions use bars; Continuous distributions are usually represented by smooth lines or curves

Ques5.What is the standard normal distribution, and why is it important?
Ans:The standard normal distribution is a special case of the normal (Gaussian) distribution that has been standardized to make analysis and comparison easier.
Why the Standard Normal Distribution is Important:
1.Central Limit Theorem (CLT): This foundational theorem in statistics states that the sampling distribution of the mean approaches a normal distribution, regardless of the population's shape, as the sample size increases.
2.Modeling Real-World Data: It models data where most values cluster around the average, such as human heights, IQ scores, or errors in scientific measurements.
3.Simplifies Probabilities: Instead of calculating complex probabilities for every normal distribution, mathematicians use one standardized "Z-table" to find area (probability) under the curve, ensuring consistent calculations.

Ques7.What is the significance of confidence intervals in statistical analysis?
Ans:Confidence intervals (CIs) signify the reliability and precision of a sample estimate by providing a range of plausible values for a population parameter, rather than a single point estimate. They quantify uncertainty, showing how sampling variation affects results, typically using a 95% confidence level to indicate that true values will fall within the range in 95% of repeated studies.

Ques9.Write a Python program to generate 1000 random numbers from a normal distribution with mean = 50 and standard deviation = 5. Compute its mean and standard deviation using NumPy, and draw a histogram to visualize the distribution.

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mean = 50
std_dev = 5
n = 1000

# Generate 1000 random numbers from a normal distribution
data = np.random.normal(mean, std_dev, n)

# Compute mean and standard deviation
sample_mean = np.mean(data)
sample_std = np.std(data)

print("Sample Mean:", sample_mean)
print("Sample Standard Deviation:", sample_std)

# Plot histogram
plt.hist(data, bins=30, color='skyblue', edgecolor='black')

# Add titles and labels
plt.title("Histogram of Normally Distributed Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Show plot
plt.show()
```

QUES10.You are working as a data analyst for a retail company. The company has
collected daily sales data for 2 years and wants you to identify the overall sales trend.
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
235, 260, 245, 250, 225, 270, 265, 255, 250, 260]
● Explain how you would apply the Central Limit Theorem to estimate the average sales
with a 95% confidence interval.
● Write the Python code to compute the mean sales and its confidence interval.
Ans:1.Treat this data as a sample from all possible daily sales
2.According to the CLT:
.The sampling distribution of the mean will be approximately normal
.This holds even if the original data isn’t perfectly normal (especially with reasonable sample size)

```python
import numpy as np
from scipy import stats

# Given data
daily_sales = [220, 245, 210, 265, 230, 250, 260, 275, 240, 255,
               235, 260, 245, 250, 225, 270, 265, 255, 250, 260]

# Convert to NumPy array
data = np.array(daily_sales)

# Sample size
n = len(data)

# Mean and standard deviation
mean = np.mean(data)
std_dev = np.std(data, ddof=1)  # sample standard deviation

# Standard error
se = std_dev / np.sqrt(n)

# 95% confidence interval using t-distribution
confidence = 0.95
t_value = stats.t.ppf((1 + confidence) / 2, df=n-1)

margin_error = t_value * se

lower_bound = mean - margin_error
upper_bound = mean + margin_error

print("Mean Daily Sales:", mean)
print("95% Confidence Interval:", (lower_bound, upper_bound))
```
